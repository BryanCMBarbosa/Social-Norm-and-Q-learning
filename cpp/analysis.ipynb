```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_all_good = pd.read_csv("1111111111111111_norm_name=All-good_runs=50.csv")
df_all_bad = pd.read_csv("0000000000000000_norm_name=All-bad_runs=50.csv")
df_half_good = pd.read_csv("0000000011111111_norm_name=half-good_runs=50.csv")
df_half_bad = pd.read_csv("1111111100000000_norm_name=half-bad_runs=50.csv")
```

```python
def plot_coop_index(df, title):
    plt.figure().set_figwidth(85)
    N = len(df.columns)
    cmap = plt.cm.get_cmap("hsv", N+1)
    medians = np.array([])
    means = np.array([])

    for c in df.columns:
        medians = np.append(medians, df[c].median())
        means = np.append(means, df[c].mean())
        for r in df[c]:
            plt.scatter(c, r, c=np.array([cmap(int(c))]), edgecolors='none', alpha=0.1)

    plt.plot(df.columns, medians, color="red", marker="o", label="Median")
    plt.plot(df.columns, means, color="blue", marker="s", label="Mean")

    plt.legend(bbox_to_anchor=(0, 1), loc='upper left', ncol=1)
    plt.tight_layout()

    spacing = 0.400
    plt.subplots_adjust(bottom=spacing)

    plt.margins(x=0.01)
    
    plt.xlabel("Strategy")
    plt.ylabel("Cooperation index")
    plt.title(title)

    plt.show()
```

```python
plot_coop_index(df_all_good, "All good - 1111111111111111")
```

```python
plot_coop_index(df_all_bad, "All bad - 0000000000000000")
```

```python
plot_coop_index(df_half_good, "Half good - 0000000011111111")
```

```python
plot_coop_index(df_half_bad, "Half bad - 1111111100000000")
```
